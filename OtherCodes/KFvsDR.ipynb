```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.read_csv('KFvsDR.csv')
print(df.head())
```

```python
font = {
#     'family' : 'normal',
#   'weight' : 'bold',
  'size' : 26}
matplotlib.rc('font', **font)

fig = plt.figure(figsize=(16,9))
ax = fig.add_subplot(111)
ax.plot(df['Update Rate'],df['KF RMSE'], 'o-', label='KF RMSE')
ax.plot(df['Update Rate'], df['ADR RMSE'], 's-', label='ADR RMSE')
ax.set_xlabel('Update Rate (ms)')
ax.set_ylabel('RMSE')
ax.set_ylim(0,8)
ax.set_xticks([0.01,0.03,0.05,0.07,0.09,0.11,0.13,0.15,0.17,0.19])
ax.set_xticklabels([10,30,50,70,90,110,130,150,170,190])
ax.legend(loc='best')
plt.show()
```

```python
df = pd.read_csv('KFvsDR Trajectory.csv')
print(df.head())
```

```python
font = {
#     'family' : 'normal',
#   'weight' : 'bold',
  'size' : 24}
matplotlib.rc('font', **font)

fig = plt.figure(figsize=(16,9))
ax = fig.add_subplot(111, projection='3d')
ax.plot(df['DRx'], df['DRz'], df['DRy'], label='ADR Estimate')
ax.plot(df['KFx'], df['KFz'], df['KFy'], label='KF Estimate')
ax.plot(df['Xr'], df['Zr'], df['Yr'], label='Real')
ax.set_xlabel('X')
ax.set_ylabel('Z')
ax.set_zlabel('Y')
ax.xaxis.labelpad=20
ax.yaxis.labelpad=20
ax.legend()
plt.show()
```
